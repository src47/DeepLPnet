```python
from helper_functions import * 
from topas_scripts import *
import numpy as np
from matplotlib import pyplot as plt
import scipy
import pandas as pd
import skued
from skued import baseline_dt

tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

```python
def processExptData(Xdata, measured_wavelength=0.7293, showPlots=False, baseline=False):
    q = angle2q(Xdata[0], lbda=measured_wavelength)
    intensity = Xdata[1]
        
    if baseline: 
        intensity = intensity - baseline_dt(intensity, wavelet = 'qshift3', level = 9, max_iter = 1000)
    
    intensity = intensity/np.max(intensity)
    intensity[intensity < 0.001] = 0
    
    intensity_interpolated_no_peak_finding = interpolate_data(q, intensity, lbda=1.54056,ml_input_size=9000)
    intensity_interpolated_no_peak_finding = intensity_interpolated_no_peak_finding/np.max(intensity_interpolated_no_peak_finding)
    
    intensity_reconstruct = find_peaks_wrapper(intensity,widths=np.linspace(5,50,80))
    intensity_interpolated_peak_finding = interpolate_data(q, intensity_reconstruct, lbda=1.54056,ml_input_size=9000)
    intensity_interpolated_peak_finding = intensity_interpolated_peak_finding/np.max(intensity_interpolated_peak_finding)
    #for i in range(len(intensity_interpolated_peak_finding)):
    #    if intensity_interpolated_peak_finding[i] < 0.05:
    #        intensity_interpolated_peak_finding[i] = 0
    
    #if showPlots:
        #plt.plot(intensity_interpolated_no_peak_finding)
        #plt.plot(intensity_interpolated_peak_finding)
        #plt.show()
    return intensity_interpolated_no_peak_finding, intensity_interpolated_peak_finding

def predictExptDataPipeline(Xdata, y_true, crystal_system, measured_wavelength=0.7293, model=None, baseline=False):

    if model == None:
        #k = "_gaussian_noise"
        #"_all_except_multiphase"
        model = tf.keras.models.load_model("models_ICSD_CSD/" + crystal_system +  "_all"
)
    intensity_interpolated_no_peak_finding, intensity_interpolated_peak_finding = processExptData(Xdata, measured_wavelength=measured_wavelength, showPlots=True, baseline=baseline)
    y_pred_peak_finding = make_prediction(np.expand_dims(np.expand_dims(intensity_interpolated_peak_finding,axis=1),axis=0), model, crystal_system)
    y_pred_no_peak_finding = make_prediction(np.expand_dims(np.expand_dims(intensity_interpolated_no_peak_finding,axis=1),axis=0), model, crystal_system)
    
    #print("Noise baseline: ",  np.max(intensity_interpolated_no_peak_finding[0:500]))
    
    #print(" ")
    #print("True LPs: ", y_true)
    #print(" ")
    
    #print("Predicted LPs Peak Finding: ", y_pred_peak_finding)
    #print("MPE (relative to peak finding)", round(mean_absolute_percentage_error(y_pred_peak_finding,y_true),3), "%")
    #print(" ")
    
    #print("Predicted LPs w/o Peak Finding: ", y_pred_no_peak_finding)
    #print("MPE (relative to w/o peak finding)", round(mean_absolute_percentage_error(y_pred_no_peak_finding,y_true),3), "%")
    #print("----------------------------------------------------------------------------------------------------------------")
    
    return y_pred_no_peak_finding
```

```python
def make_template(crystal_system, y_pred, bound, dataName, fileoutName, measured_wavelength=0.7293):
    if crystal_system == "cubic":
        predicted_a, lower_bound_a, upper_bound_a, predicted_V, lower_bound_V, upper_bound_V = pointAndBoundsCubic(y_pred[0], bound=bound)
        return write_template_cubic(dataName=dataName, V=predicted_V, minV=lower_bound_V, maxV=upper_bound_V, a=predicted_a, mina=lower_bound_a, maxa=upper_bound_a, fileOut=fileoutName)
    
```

```python
crystal_system = 'cubic'
dataName = "BL21Robot_0001_Rietveld_X_Difference"

try:
    os.mkdir("data_expt/SSRL_samples_run/" + dataName)
except:
    pass

bounds = [0.1, 0.2, 0.5, 1]

for bound in bounds:
    
    Xdata = pd.read_csv("data_expt/" + dataName + ".xy",sep='\t',header=None)
    y_true = np.array([4.156826, 4.156826, 4.156826])
    y_pred = predictExptDataPipeline(Xdata, y_true, crystal_system=crystal_system)
    fileoutName = os.path.join("data_expt", "SSRL_samples_run", dataName, dataName + "_" + str(bound) + ".inp")
    lines = make_template(crystal_system='cubic', y_pred=y_pred, bound=bound, dataName=dataName, fileoutName=fileoutName)
```
